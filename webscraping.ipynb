```python
import requests 
url = "https://fr.wikipedia.org/wiki/Championnat_de_France_de_football_2019-2020"

# Renvoie le code source de la page en question
request_text = requests.get(
    url, 
    headers={"User-Agent": "Python for data science tutorial"}
    ).content
```

```python
import bs4

# Renvoie une page HTML, on met le "lxml" à chaque fois
page = bs4.BeautifulSoup(request_text, "lxml")
print(page)
```

```python
print(page.find("table"))
print(page.find_all("table")) # Sous la forme d'une liste
print("Il y a", len(page.find_all("table")), "éléments dans la page qui sont des tables")
```

```python
# Exercice : trouver l'ensemble des participants de la ligue 1

# Etape 1 : on récupère le tableau des participants
tableau = page.find("table", {'class': 'wikitable sortable'})

# Etape 2 : on récupère les lignes
corps_tableau = tableau.find('tbody')
lignes = corps_tableau.find_all("tr")

# Etape 3 et 4 : on récupère les colonnes de chaque ligne, dont on extrait le texte
matrice = []
for ligne in lignes : 
    colonnes = ligne.find_all('td')
    colonnes = [cellule.text.strip() for cellule in colonnes]
    matrice.append(colonnes)

# Etape 5 : on en fait un dictionnaire
dictionnaire = {}

for ligne in matrice :
    if len(ligne) > 0 :
        dictionnaire[ligne.pop(0)] = ligne
print(dictionnaire)

# Etape 6 : transformation en dataframe
import pandas as pd
dataframe = pd.DataFrame.from_dict(dictionnaire, orient='index')

print(dataframe.head())

# Etape 7 : Récupérer les en-têtes du tableau
en_tetes = [cellule.get_text(separator = " ").strip().title() for cellule in lignes[0].find_all('th')]
print(en_tetes)

dataframe.columns = en_tetes[1:]
dataframe.head()
```
